### Si BCBQ

```python
import pandas as pd
import data
import torch
import numpy as np


energy_4s_bq = {'mix_name': [], 'voices': [], 'eval_frame': [], 'energy': []}

# bc_val = data.BCBQDataSets(data_set='BC',
#                             validation_subset=True,
#                             conf_threshold=0.4,
#                             example_length=64000,
#                             n_sources=4,
#                             random_mixes=False,
#                             return_name=True,
#                             allowed_voices='satb',
#                             f0_from_mix=False,
#                             cunet_original=False,
#                             cuesta_model=False,
#                             cuesta_model_trainable=False)

bq_val = data.BCBQDataSets(data_set='BQ',
                        validation_subset=True,
                        conf_threshold=0.4,
                        example_length=64000,
                        n_sources=4,
                        random_mixes=False,
                        return_name=True,
                        allowed_voices='satb',
                        f0_from_mix=False,
                        cunet_original=False,
                        cuesta_model=False,
                        cuesta_model_trainable=False
                        )

test_set = torch.utils.data.ConcatDataset([bq_val]) 

device = 'cuda' if torch.cuda.is_available() else 'cpu'

energy_total = []
for idx in range(len(test_set)):
    # Load batch of 4 1-second frames
    d = test_set[idx]

    mix = d[0].to(device)
    f0_hz = d[1].to(device)
    target_sources = d[2].to(device)
    name = d[3]
    voices = d[4]


    mix = mix.unsqueeze(0)
    target_sources = target_sources.unsqueeze(0)
    f0_hz = f0_hz[None, :, :]

    batch_size, n_samples, n_sources = target_sources.shape

    n_fft_metrics = 512
    overlap_metrics = 0.5
    

    target_sources = target_sources.transpose(1, 2)  # [batch_size, n_sources, n_samples]
    target_sources = target_sources.reshape((batch_size * n_sources, n_samples))
    
    n_source, n_samples = target_sources.shape
    n_eval_frames = int(np.ceil(n_samples / 16000))
    
    energy_batch = []
    for source in range(n_source):
        for n in range(n_eval_frames):
            energy_4s_bq['mix_name'].append(name)
            energy_4s_bq['voices'].append(voices[source])
            energy_4s_bq['eval_frame'].append(n)
            energy_4s_bq['energy'].append(torch.sum(torch.square(target_sources[source, n*16000: (n+1)*16000])).cpu().numpy())
            

energy_4s_bq = pd.DataFrame(energy_4s_bq)
```

```python
energy_to_drop = []
for i, energy in enumerate(energy_4s_bq["energy"]):
    if energy < 10:
        for j in range(4):
            energy_to_drop.append(i + j * len(energy_4s_bq["energy"]))
```

```python
df = df.drop(energy_to_drop)

means = df.mean(axis=0, skipna=True, numeric_only=True)
medians = df.median(axis=0, skipna=True, numeric_only=True)
stds = df.std(axis=0, skipna=True, numeric_only=True)

print('SI-SDR [mean, median, std]:', means['SI-SDR'], medians['SI-SDR'], stds['SI-SDR'])
print('sp_SNR [mean, median, std]:', means['sp_SNR'], medians['sp_SNR'], stds['sp_SNR'])
print('sp_SI-SNR [mean, median, std]:', means['sp_SI-SNR'], medians['sp_SI-SNR'], stds['sp_SI-SNR'])
print('mel_cep_dist [mean, median, std]:', means['mel_cep_dist'], medians['mel_cep_dist'], stds['mel_cep_dist'])
```
